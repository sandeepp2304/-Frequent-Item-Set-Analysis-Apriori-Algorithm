```python
# # # # # # # # #  Code for Task 1 # # # # # # # # #


# # # # # # # # #  Code for Task 1 # # # # # # # # #
df_item = pd.read_csv('dataset_group.csv', delimiter = ',',header = None);

df_item.columns = ['Date', 'Bill', 'Item']
df_item['Date'] = df_item['Date'].apply(dateutil.parser.parse, dayfirst=True);

df_items = df_item.groupby(['Date','Bill'])['Item'].unique()
```

```python
df_items.head()
```

```python
df = []
for i,j in df_items.iteritems():
    df.append(j)
```

```python
df=pd.DataFrame(df)

df.to_csv('dataset.csv', index=False,header = 0)

print('First 7 transactions of the dataset.csv');
df.head(7)
```
